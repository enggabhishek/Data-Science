```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
with open('amazon_jobs_dataset.csv',encoding='utf-8') as obj:
    data=pd.read_csv(obj)
    x=[]
    y=[]
    date=data['Posting_date']
    year=date.str.split(', ', expand=True)[1]
    freq=year.value_counts(ascending=True)
    freq.sort_index(inplace=True)
    x=freq.index
    y=freq
    plt.plot(x,y,marker='o')
    plt.xlabel("No. of Job postings")
    plt.ylabel("Year")
    plt.legend()
    plt.grid()
    plt.show()
    for i in range(len(x)):
        print(x[i],y[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
with open('amazon_jobs_dataset.csv',encoding='utf-8') as obj:
    data=pd.read_csv(obj)
    x=[]
    y=[]
    date=data['Posting_date']
    month=date.str.split(' ', expand=True)[0]
    freq=month.value_counts(ascending=True)
    x=freq.index
    y=freq
    plt.plot(x,y,marker='o')
    plt.xlabel("No. of Job postings")
    plt.ylabel("Month")
    plt.legend()
    plt.xticks(rotation=90)
    plt.grid()
    plt.show()
    for i in range(len(x)):
        print(x[i],y[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
with open('amazon_jobs_dataset.csv',encoding='utf-8') as obj:
    data=pd.read_csv(obj)
    x=[]
    y=[]
    location=data['location']
    country=location.str.split(', ', expand=True)
    arr=country[0]=='IN'
    city=country[arr][2]
    y=city.value_counts(ascending=False)
    x=y.index
    plt.axis("equal")
    plt.pie(y,labels=x,autopct='%.2f%%')
    plt.show()
    perc=np.true_divide(y, y.sum(axis=0))*100
    for i in range(len(x)):
        print(x[i],format(perc[i],'.2f'))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
with open('amazon_jobs_dataset.csv',encoding='utf-8') as obj:
    data=pd.read_csv(obj)
    basic_Qualification=data['BASIC QUALIFICATIONS']
    year=data.Posting_date.str.split(', ', expand=True)[1]
    bq=np.array(list(zip(year,basic_Qualification)))
    for i in range(len(bq)):
        if 'Java' in bq[i][1] or 'java' in bq[i][1]:
            bq[i][1]=1
        else:
            bq[i][1]=0
    bq=bq[bq[:,1]=='1']
    bq=bq[:,0]
    y=np.unique(bq)
    x=[]
    for i in y:
        x.append(len(bq[bq==i]))
    plt.scatter(y,x,edgecolors='black',s=100)
    plt.show()
    for i in range(len(y)):
        print(y[i],x[i])
```
